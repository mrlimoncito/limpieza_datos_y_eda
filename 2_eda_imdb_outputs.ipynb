### 10. Inspecciona si hay columnas que tengan valores `NA`

```python
print("Valores NA por columna:")
display(df.isnull().sum())
```

### 11. Imprime el número total de valores NA que haya en cada columna. Hazlo primero para la columna `Metascore` y después utiliza un ciclo for para hacerlo para todas las columnas

```python
print("Valores NA en la columna 'Metascore':")
print(df['Metascore'].isnull().sum())

print("\nValores NA en todas las columnas (usando un ciclo for):")
for column in df.columns:
    print(f"Columna '{column}': {df[column].isnull().sum()} NA valores")
```

### 15. Crea un histograma de la variable Metascore. Utiliza 10 cubetas

### 17. Vuelve a describir el dataframe y observa la media de la variable `Ratings`

```python
print("Estadísticos descriptivos del DataFrame (después de limpiar):")
display(df_cleaned.describe())
print("Media de la variable 'Rating':")
print(df_cleaned['Rating'].mean())
```

### 18. Calcula este promedio con Numpy y después con un método de Pandas

```python
rating_mean_numpy = np.mean(df_cleaned['Rating'])
print(f"Promedio de 'Rating' con NumPy: {rating_mean_numpy:.4f}")

rating_mean_pandas = df_cleaned['Rating'].mean()
print(f"Promedio de 'Rating' con Pandas: {rating_mean_pandas:.4f}")
```

### 19. Obten los valores únicos de la variable Rating y después ordénalos de menor a mayor

```python
unique_ratings = df_cleaned['Rating'].unique()
print("Valores únicos de 'Rating':")
print(unique_ratings)

sorted_unique_ratings = np.sort(unique_ratings)
print("\nValores únicos de 'Rating' ordenados de menor a mayor:")
print(sorted_unique_ratings)
```

### 20. Observa los ratings que te interesen y ahora filtra el dataframe con ese rating para ver cuáles son las películas con dicho rating

```python
rating_interes = 9.0 # Puedes cambiar este valor al rating que te interese
peliculas_filtradas = df_cleaned[df_cleaned['Rating'] == rating_interes]

print(f"Películas con rating {rating_interes}:")
display(peliculas_filtradas)
```

### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con essos valores

```python
rating_counts = df_cleaned['Rating'].value_counts().reset_index()
rating_counts.columns = ['rating', 'frequency']

print("Frecuencia de cada valor de 'Rating':")
display(rating_counts)
```

### 22. Ordena el nuevo dataframe por la variable `rating`

```python
rating_counts_sorted = rating_counts.sort_values(by='rating', ascending=True)

print("Frecuencia de Rating ordenada por 'rating':")
display(rating_counts_sorted)
```

### 24. Crea la matriz de correlación del dataframe de películas

```python
correlation_matrix = df_cleaned.corr(numeric_only=True)

print("Matriz de correlación del DataFrame de películas:")
display(correlation_matrix)
```

### 25. Grafica la matriz de correlación utilizando `matshow()`

```python
plt.figure(figsize=(10, 8))
plt.matshow(correlation_matrix, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(correlation_matrix.shape[1]), correlation_matrix.columns, rotation=90)
plt.yticks(range(correlation_matrix.shape[1]), correlation_matrix.columns)
plt.title('Matriz de Correlación de Películas', pad=20)
plt.tight_layout()
plt.show()
```
